```python
# Initialize a list to hold all DataFrames
df_list = []

# The pattern that matches a job listing
pattern = r'\{"id":".*?"jobstreet:s_vi":""\}\}\}'

# Loop over each URL
for url in urls:
    # Send HTTP request to URL and save the response from server in a response object called r
    r = requests.get(url)

    # Create a BeautifulSoup object and specify the parser
    soup = BeautifulSoup(r.text, 'html.parser')

    # Use the 'find_all' function to find all script tags in the HTML
    script_tags = soup.find_all('script')

    # Combine all the script tags into a single string
    all_scripts_str = ' '.join(str(tag) for tag in script_tags)

    # Use a regular expression to find all matches in the combined script string
    matches = re.findall(pattern, all_scripts_str)

    # Initialize an empty list to store the jobs data
    jobs_data = []

    # Loop through the matches
    for match in matches:
        # Try to parse the JSON data
        data = json.loads(match)
        # Append the data to the jobs list
        jobs_data.append(data)

    # Convert the list of jobs into a DataFrame
    df = pd.DataFrame(jobs_data)

    # Append the DataFrame for this URL to the df_list
    df_list.append(df)

# Concatenate all the DataFrames in the df_list
all_jobs_data = pd.concat(df_list, ignore_index=True)
```

Let's start with the first step: parsing the JSON columns. We'll start with the salaryRange column, extracting minSalary, maxSalary, and currency columns. We'll also need to handle cases where these fields might not be present in the JSON object.

```python
# Define a function to parse the salaryRange column
def parse_salary_range(salary_range):
    if pd.isna(salary_range) or salary_range == "None":
        return np.nan, np.nan, np.nan, np.nan, np.nan
    else:
        salary_range_dict = salary_range
        min_salary = salary_range_dict.get('min')
        max_salary = salary_range_dict.get('max')
        currency = salary_range_dict.get('currency')
        period = salary_range_dict.get('period')
        term = salary_range_dict.get('term')
        return min_salary, max_salary, currency, period, term

# Apply the function to the salaryRange column
all_jobs_data[['minSalary', 'maxSalary', 'currency', 'period', 'term']] = all_jobs_data['salaryRange'].apply(parse_salary_range).apply(pd.Series)

all_jobs_data.info()
```

```python
# Define a function to parse the companyMeta column
def parse_company_meta(company_meta):
    if pd.isna(company_meta) or company_meta == "None":
        return np.nan, np.nan, np.nan, np.nan, np.nan
    else:
        company_meta_dict = company_meta
        company_meta_id = company_meta_dict.get('id')
        ad_id = company_meta_dict.get('advertiserId')
        isprivate = company_meta_dict.get('isPrivate')
        company_meta_name = company_meta_dict.get('name')
        company_meta_slug = company_meta_dict.get('slug')
        return company_meta_id, ad_id, isprivate, company_meta_name, company_meta_slug

# Apply the function to the companyMeta column
all_jobs_data[['companyMetaId', 'AdId', 'isPrivate', 'companyMetaName', 'companyMetaSlug']] = all_jobs_data['companyMeta'].apply(parse_company_meta).apply(pd.Series)

all_jobs_data.info()
```

```python
# Define a function to parse the employmentTypes column
def parse_employment_types(employment_types):
    if pd.isna(employment_types) or employment_types == "None":
        return np.nan, np.nan
    else:
        employment_types_list = employment_types
        if len(employment_types_list) > 0:
            first_dict = employment_types_list[0]
            emp_code = first_dict.get('code')
            emp_name = first_dict.get('name')
        else:
            emp_code = emp_name = np.nan
        return emp_code, emp_name

    
# Apply the function to the employmentTypes column
all_jobs_data[['employmentCode', 'employmentName']] = all_jobs_data['employmentTypes'].apply(parse_employment_types).apply(pd.Series)

all_jobs_data.info()
```

```python
# Convert lists in 'sellingPoints' to strings
all_jobs_data['sellingPoints'] = all_jobs_data['sellingPoints'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# Print the first few rows of the updated 'sellingPoints' column
print(all_jobs_data['sellingPoints'].head())
```

```python
# Define a function to extract location names from a list of dictionaries
def extract_location_names(location_list):
    names = [location['name'] for location in location_list]
    return ', '.join(names)

# Apply the function to the 'locations' column
all_jobs_data['location_names'] = all_jobs_data['locations'].apply(extract_location_names)

# Print the first few rows of the updated DataFrame
print(all_jobs_data['location_names'].head())

all_jobs_data.info()
```
